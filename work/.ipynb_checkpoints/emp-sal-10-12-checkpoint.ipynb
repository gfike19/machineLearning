```python
import pandas as pd
import numpy as np

empData = pd.read_csv("C:/Users/aerot/coding/machineLearning/Week3/Assignments/emp_sal.csv")
empDF = pd.DataFrame(empData)
```

```python
# •	Find the net salary after 30% deduction of tax
empDF = empDF.rename(columns={" Salary" : "Salary"})
sals = np.array(empDF["Salary"])
netSals = []
for val in sals:
    intSal = int(val[1:]) * .3
    netSals.append(intSal)
print(netSals)
```

```python
# •	remove all rows having null value
empDF.dropna()
print(empDF)
```

```python
# •	Create another col Int_Sal where salary is stored as integer type
intSals = []
for val in sals:
    intSal = int(val[1:])
    intSals.append(intSal)
empDF["Int Salary"] = intSals
empDF
```

```python
# •	Find the name of the employee having maximum salary
empDF = empDF.sort_values(by=["Salary"])
empDF.head(1)
```
