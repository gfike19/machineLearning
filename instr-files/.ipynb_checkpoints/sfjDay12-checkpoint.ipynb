```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('/Users/Ganesh/Desktop/SFJ/puppy_dg.jpg')
```

```python
# im = cv2.imread("/Users/Ganesh/Desktop/SFJ/puppy_dg.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
im_flat = np.divide(img, 255)
```

```python
im_flat.shape
```

```python
X = im_flat.reshape(-1, 3)
```

```python
from sklearn.cluster import KMeans
k=4
kmeans = KMeans(n_clusters=k, random_state=0, verbose=0, max_iter=5)
```

```python
kmeans.fit(X)
```

```python
# complete then send
```

```python
X_recovered = kmeans.cluster_centers_[kmeans.labels_ , :].reshape(img.shape)
```

```python
X_recovered = X_recovered * 255
```

```python
from matplotlib import pyplot
fig, ax = pyplot.subplots(1, 2, figsize=(8, 4))
ax[0].imshow(im_flat)
ax[0].set_title('Original')
ax[0].grid(False)

ax[1].imshow(X_recovered/255)
ax[1].set_title('Compressed, with %d colors' % k)
ax[1].grid(False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data[:, :4]
y = iris.target
```

```python
X[:10]
```

```python
ad = pd.read_csv('/Users/Ganesh/Desktop/SFJ/Data/Social_Network_Ads.csv')
ad.head()
```

```python
ad.Purchased.unique()
```

```python
X = ad.iloc[:,[2,3]].values
y = ad.iloc[:,4].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train.shape
```

```python
classifier = SVC()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred) * 100
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('darkred', 'lightgreen'))(i), label = j)
```
